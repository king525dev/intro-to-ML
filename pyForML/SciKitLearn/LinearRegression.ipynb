# Simple Linear Regression Algorithm

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Importing Data

df = pd.read_csv("data/testData.csv")
df.head()
```

```python
df.shape
```

```python
# Features Matrix

X = df.loc[:, ["x"]].values
```

```python
X
```

```python
# Arranging Results matrix

y = df.loc[:, 'y'].values
```

```python
# Graph Data

plt.style.use('fivethirtyeight')
plt.plot(X, y)
```

```python
# Regression Model

reg = LinearRegression(fit_intercept=True)
reg.fit(X,y)
```

```python
# Scoring the Algorithm

score = reg.score(X, y)
print(score)
```

```python
# Importing test data

df = pd.read_csv("data/testData.csv")
df.head()
```

```python
# Comparing algorithm and function

Xtest = df.loc[:, ["x"]].values
Ytest = reg.predict(Xtest)
fun = df.loc[:, ["y"]].values

dater = Ytest,fun


colNames = [
     "Predicted Values",
     "Actual Values"
]
```

```python
# Equation of the Line of best fit
m = reg.coef_[0]
b = reg.intercept_

print("formula: y = {:.2f}x + {:.2f}".format(m, b) )
```

```python
# Plotting graph

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10,7));

ax.scatter(X, y, color='black');
ax.plot(X, reg.predict(X), color='red',linewidth=3);
ax.grid(True,
          axis = 'both',
          zorder = 0,
          linestyle = ':',
          color = 'k')
ax.tick_params(labelsize = 18)
ax.set_xlabel('x', fontsize = 24)
ax.set_ylabel('y', fontsize = 24)
ax.set_title("Linear Regression Line with Intercept y = {:.2f}x + {:.2f} (R2 = {:.2f})".format(m, b, score), fontsize = 16 )
fig.tight_layout()
fig.savefig('linearregression', dpi = 500)
```

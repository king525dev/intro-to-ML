```python
import numpy as np

# Creating a 1D array
arr1 = np.array([1, 2, 3, 4])

# Creating a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print(arr1)
print(arr2)
```

```python
# Accessing elements in a 1D array
print(arr1[0])  # Output: 1
print(arr1[3])  # Output: 4

# Accessing elements in a 2D array
print(arr2[0, 1])  # Output: 2 (first row, second column)
print(arr2[1, 2])  # Output: 6 (second row, third column)
```

```python
# Reshaping a 1D array into a 2D array
arr3 = np.array([1, 2, 3, 4, 5, 6])
reshaped_arr = arr3.reshape(2, 3)

print(reshaped_arr)
```

```python
arr4 = np.array([10, 20, 30, 40])

# Element-wise addition
sum_arr = arr1 + arr4

# Element-wise subtraction
diff_arr = arr4 - arr1

# Element-wise multiplication
prod_arr = arr1 * arr4

# Element-wise division
div_arr = arr4 / arr1

print("Sum:", sum_arr)
print("Difference:", diff_arr)
print("Product:", prod_arr)
print("Division:", div_arr)
```

```python
import pandas as pd

data = {
     'Name': ['Alice', 'Bob', 'Charlie'],
     'Age': [25, 30, 35],
     'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Accessing a column
print(df['Name'])

# Accessing multiple columns
print(df[['Name', 'Age']])

# Accessing a row by index
print(df.iloc[0])

# Accessing a row by label
print(df.loc[0])
```

```python
data = {
     'Category': ['A', 'A', 'B', 'B'],
     'Values': [10, 15, 10, 25]
}
df = pd.DataFrame(data)

# Group by Category and sum the Values
grouped_df = df.groupby('Category').sum()
print(grouped_df)
```

```python
data = {
     'Name': ['Alice', 'Bob', 'Charlie'],
     'Age': [25, None, 35],
     'City': ['New York', None, 'Chicago']
}
df = pd.DataFrame(data)

# Drop rows with missing values
df_dropna = df.dropna()
print(df_dropna)

# Fill missing values with a specific value
df_fillna = df.fillna({'Age': 30, 'City': 'Unknown'})
print(df_fillna)

print(df.describe())
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Seaborn Bar Plot')
plt.show()
```

```python
#
#     TEXT ANLYSER
#

from collections import Counter
import matplotlib.pyplot as plt


def anaylseText(text):
     words = text.split()
     wordFrequency = Counter(words)
     uniqueWords = set(words)

     return [wordFrequency, uniqueWords]

def outputAnalysis(wordFrequency, uniqueWords):
     return f""" 
     Word Frequency: {wordFrequency}
     Unique Words: {uniqueWords}
     """

def plotAnalysis(wordFrequency, uniqueWords):
     freq = wordFrequency.values()

     plt.bar(uniqueWords, freq)
     plt.xlabel('Words')
     plt.ylabel('Frequency')
     plt.title('Frequency of Words')
     plt.show()

userInput = input("Input text to be analysed: ")
output = anaylseText(userInput)

print(outputAnalysis(output[0], output[1]))
plotAnalysis(output[0], list(output[1]))
```
